```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('Datos.xlsx', index_col = 'Nombre');
```

```python
df.head()
```

```python
# Media
X1_media = df['X1'].mean()
X2_media = df['X2'].mean()
```

```python
# Maximo
X1_max = df['X1'].max()
X2_max = df['X2'].max()
```

```python
df['X1_cat'] = pd.cut(df['X1'], bins = [0,X1_media,X1_max], labels = ['Pobre','Rico'])
```

```python
df['X2_cat'] = pd.cut(df['X2'], bins = [0,X2_media,X2_max], labels = ['Joven','Viejo'])
```
