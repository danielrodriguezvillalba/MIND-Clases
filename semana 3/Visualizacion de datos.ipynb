```python
# Reemplazar espacios por rayas al piso
df.columns = df.columns.str.replace(' ','_')
```

```python
#clasificar por numero la categoria 
df['color_cod'] = df.color.cat.codes
```

```python
df.color = pd.Categorical(df.color)
```

```python
df.head()
```

```python
# Graficar diagrama de dispercion
df.plot.scatter(x='fixed_acidity',y='volatile_acidity')
```

```python
df.plot.scatter(x='fixed_acidity',y='volatile_acidity',c='color_cod')
```

```python
df.plot.scatter(x='fixed_acidity',y='volatile_acidity',c='color_cod',colormap='prism')
# Guarda la visualizacion
plt.savefig('scatter.png')
```
